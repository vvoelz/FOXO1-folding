```python
import pandas as pd

# df = pd.DataFrame(data=dic_matrix)
# df.to_excel("/home/leiqian/Documents/GROMACS_FOXO1/Graduation/images/figures/fig_si_2/matrix_plots.xlsx")

df = pd.read_excel("/home/leiqian/Documents/GROMACS_FOXO1/Graduation/images/figures/fig_si_2/matrix_plots.xlsx", \
                   sheet_name=0, engine="openpyxl")

df = df.iloc[:, 1:]

row_num = len(df)
col_num = len(df.columns)
col_name = list(df.columns)

dic_matrix_new = {}

for i in range(col_num):
    col_name_X = col_name[i]
    col_list_X = []
    
    for j in range(row_num):
        
        item_str = df.iloc[j,i].strip("'[").strip("]'").split()
        item_list = [float(x[:-1]) for x in item_str]
        
        col_list_X.append(item_list)
        
    dic_matrix_new.update({col_name_X: col_list_X})
    
print(dic_matrix_new)
```

```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import itertools

titles_names = list(dic_matrix_new.keys())
norm1 = matplotlib.colors.LogNorm(vmin=0.01, vmax=0.99)
cmap1 = "rainbow"

grp1 = [0, 1, 2, 3] # modified
grp2 = [0, 1, 2, 3]
pairs = list(itertools.product(grp1, grp2))
# print(pairs)

fig, axes = plt.subplots(ncols=4, nrows=4, figsize=(22,17.3)) # modified

for i in range(0, 16): # modified

    t = titles_names[i]
        
    index_0 = pairs[i][0]
    index_1 = pairs[i][1]
    ax = axes[index_0, index_1]

    cax = ax.matshow(dic_matrix_new[t], cmap=plt.get_cmap(cmap1), norm=norm1)

    ax.text(0.5, -0.1, t, horizontalalignment='center', transform=ax.transAxes, size=15)

    cbar = fig.colorbar(cax, ax=axes[index_0, index_1], orientation="vertical", shrink=0.65)  
    cbar.ax.tick_params(labelsize=13) 

plt.tight_layout()   

plt.savefig('/home/leiqian/Documents/GROMACS_FOXO1/Graduation/images/figures/fig_si_2/SI_Figure2.png', facecolor='white', \
            dpi=600) # modified 

plt.show()
```
